#원형 차트 (파이 차트)

## 특징

- 비교적 많이 사용되지 않음
- 데이터를 깔끔하게 보여주는 것이 특징, 점유율 등 상대적으로 큰 부분을 강조할 때 사용, 갯수가 많으면 바차트로 설계하는게 합리적임 
- 데이터의 종류가 적을 때, 크기를 비교시 사용함 

```
# 데이터는 아이리스(붓꽃) 사용 
import sklearn.datasets 

# 원본
ds = sklearn.datasets.load_iris()
type(ds), ds['data'].shape, ds['feature_names'], ds['target_names']
```

```
# 기본 제공된 데이터를 df로 생성 
df = pd.DataFrame( ds['data'], columns = ds['feature_names'])
df.head(2)
```

```
# 정답 컬럼을 추가하시오
# 컬럼명은 'species'
# 단, 정답 3개는 문자열이 아닌 0, 1, 2 인 분류변수로 변경하여 
# 'setosa' => o
ds['target'], len(ds['target'])
```

```
df['specics'] = ds['target']
df.head(2)
```

```
m = list( zip(*list( enumerate(ds['target_names']))))
m
```

```
df['species2'] = [ m[1][n] for n in ds['target'] ]
df.head(2)
```

```
df2 = df[:]
del df2['specics']
df2.head(2)
```

```
# 종별로 독립변수의 합산
sums_by_species = df2.groupby('species2').sum()
sums_by_species
```

```
sbs_sepal_len = sums_by_species[ sums_by_species.columns[0] ]
sbs_sepal_len, type(sbs_sepal_len)
```

```
# pandas에서 직접 ploting 함 
sbs_sepal_len.plot( kind = 'pie', fontsize=10 )
# 데이터 이름
plt.ylabel( 'check' ) #, horizentalalignment = 'right' )
plt.title('iris ' + sums_by_species.columns[0] )
```

```
sums_by_species.plot( kind = 'pie', subplots=True, layout = (2,2), legend=False )
```

#상자 그림 (박스 플롯)

- 데이터가 어느쪽에 몰려있는지 박스 형태로 표현

```
plt.figure( figsize = (8, 6) )
sns.boxplot( x='day', y='total_bill', data=tips )
plt.show()
```

# 산포 로그

- 산포도로 데이터의 분포를 보는데, 완전 몰려있다
- 그래서 값들의 퍼짐이 구분이 안된다
- 로그 처리를 해서 퍼뜨려서 확인할때 사용함 

```
# 데이터 로드 
bs = sklearn.datasets.load_boston()
bs['data'].shape, bs.feature_names, type(bs)
```

```
# 데이터 프레임
df = pd.DataFrame( bs.data, columns = bs.feature_names)
df.head(2)
```

```
# 정답추가 
df['house_price'] = bs['target']
df.head(2)
```

```
# x축은 보스톤에서 인구 1인당 발생되는 범죄율
# y축은 집값 24.0 만달러 
# 산포도로 그리시오 

df.plot( kind = 'scatter', x='CRIM', y='house_price')
# 범죄율이 0-10 구간이 너무 많아서 구분이 안간다. => 값을 로그 처리해서 분산시키자 
```

```
# 산포로그 ---
df.plot( kind='scatter',  x='CRIM', y='house_price', logx=True)
```

#히트맵

- 데이터가 많으면 산포도가 지저분해진다 
- 시각화가 비효율화된다 
- 데이터의 전체적인 기세, 밀도, 양상을 파악
- 구역을 밀도로 표현

```
flights = sns.load_dataset('flights')
flights.shape
```

```
flights
```

```
# 데이터를 살펴보니, 여러 컬럼에서 반복적/주기적으로 데이터 값이 보인다 
# 피벗을 통해서 숨겨진 정보를 획득
flights_piv = flights.pivot('month', 'year', 'passengers')
flights_piv
```

```
# 히트맵을 통해서 고객이 어느 방향으로 증가하고 있는지 
# annot : 수치를 세팅, 보여라 
# fmt : 수치의 타입, 정수 
sns.heatmap( flights_piv, annot = True, fmt = '0' )
```
